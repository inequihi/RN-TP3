# Data Augmentation
Generamos nuevas imagenes a partir de las actuales, por ejemplo rotando la imagen.
Importante con data augmentation **reemplazamos** las imagenes de train por un batch nuevo, transformado aleatoriamente .
La idea es minimizar overfitting

```python
train_datagen = ImageDataGenerator(
    rotation_range=10,       # Rotar im√°genes hasta 180 grados
    width_shift_range=0.1,   # Desplazamiento horizontal hasta el 20% del ancho de la imagen
    height_shift_range=0.1,  # Desplazamiento vertical hasta el 20% de la altura de la imagen
    shear_range=0.2,         # Aplicar operaciones de corte
    zoom_range=0.1,          # Aplicar zoom
    horizontal_flip=True,    # Voltear im√°genes horizontalmente
    vertical_flip = False,   # Voltear im√°genes verticalmente
    fill_mode='nearest',      # Rellenar p√≠xeles vac√≠os
)

#Ajuste de generadores a datos normalizados
train_aug_images = train_datagen.flow(
                                x_train, 
                                y_train_fine,
                                batch_size=batchSize, 
                                shuffle = False)

x_train_aug = []
y_train_fine_aug = []

# Cant batches por epoch
num_batches = len(x_train) // batchSize
if len(x_train) % batchSize != 0:
    num_batches += 1
    
for i in range(num_batches):
    batch_images, batch_labels = next(train_aug_images)
    x_train_aug.append(batch_images)
    y_train_fine_aug.append(batch_labels)
    
# Concatenate all batches to get the full augmented dataset
x_train_aug = np.concatenate(x_train_aug, axis=0)
y_train_fine_aug = np.concatenate(y_train_fine_aug, axis=0)

print("Augmented images shape:", x_train_aug.shape)
print("Augmented labels shape:", y_train_fine_aug.shape)

# Visualizar algunas im√°genes originales y aumentadas
fig, axes = plt.subplots(2, 5, figsize=(10, 4))
fig.suptitle('Original y Aumentada', fontsize=16)

for i in range(5):
    # Mostrar im√°genes originales
    ax = axes[0, i]
    ax.imshow(x_train[i]) # Convertir a uint8 si es necesario
    index_label = y_train_fine[i]
    label_fine = labels_fine[index_label]
    ax.set_title('Original\n{}'.format(label_fine), fontsize=8)
    ax.axis('off')

    # Mostrar im√°genes aumentadas
    ax = axes[1, i]
    augmented_image = x_train_aug[i]  # Convertir a uint8 si es necesario
    ax.imshow(augmented_image)
    index_label = y_train_fine_aug[i]
    label_fine = labels_fine[index_label]
    ax.set_title('Augmentated\n{}'.format(label_fine), fontsize=8)
    ax.axis('off')

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
## Concateno data
print("x_train size: {}".format(x_train.shape))
print("y_train_fine size: {}".format(y_train_fine.shape))


print("x_train_aug shape:", x_train_aug.shape)
print("y_train_aug shape:", y_train_fine_aug.shape)

x_train_combined = np.concatenate((x_train, x_train_aug), axis=0)
y_train_fine_combined = np.concatenate((y_train_fine, y_train_fine_aug), axis=0)

print("Combined x_train shape:", x_train_combined.shape)
print("Combined y_train shape:", y_train_fine_combined.shape)

# Ensure the first dimension is the same for both arrays
assert x_train_combined.shape[0] == y_train_fine_combined.shape[0]

# Create a permutation of indices
indices = np.random.permutation(x_train_combined.shape[0])

# Shuffle the arrays using the permutation
x_train_combined_shuffled = x_train_combined[indices]
y_train_fine_combined_shuffled = y_train_fine_combined[indices]

print("Combined x_train shape shuffled:", x_train_combined.shape)
print("Combined y_train shape shuffled:", y_train_fine_combined.shape)


# Visualizar algunas im√°genes originales y aumentadas concatenadas
fig = plt.figure(figsize=(8,8))
for i in range(40):
    axis = fig.add_subplot(4, 10, 1+i, xticks = [], yticks = [])
    plt.imshow(x_train_combined_shuffled[i])
    index_label = y_train_fine_combined_shuffled[i]
    label_fine = labels_fine[index_label]
    axis.set_title("{}".format(label_fine), fontsize=8)

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

## Modelo Secuencial Con Capas Convolucionales

```python

## Agrego 3 capas Conv2D con MaxPooling2D para downsampling 
## Cada capa de convolucional usa activacion ReLU y para normal inicializacion Glorot
## Despues de cada bloque convolucional aplico Dropout y BatchNormalization

## Kernel size (1, 1), (3, 3), (5, 5), or (7, 7)

## Dimension de salida Wo x Hx x K
# W: Dimension original
# F: tama√±o del filtro - si es recatngular en Wo pongo una y en Ho otra
# p: padding
# S: strides

# Wo = [ W - F + p( (F-1) / 2 )  / S ] + 1

model = Sequential()

model.add(Conv2D(256,(3,3),padding='same',input_shape=(32,32,3)))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(Conv2D(256,(3,3),padding='same'))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Dropout(0.2))
 
model.add(Conv2D(512,(3,3),padding='same'))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(Conv2D(512,(3,3),padding='same'))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Dropout(0.2))

model.add(Conv2D(512,(3,3),padding='same'))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(Conv2D(512,(3,3),padding='same'))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Dropout(0.2))

model.add(Conv2D(512,(3,3),padding='same'))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(Conv2D(512,(3,3),padding='same'))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Dropout(0.2))

model.add(Flatten())


# Flatten va desps de capas convolucionales
model.add(Flatten(input_shape=(32,32,3)))

## Adding dense layers  SOLO DOS MENOS 
model.add(Dense(1024, activation="relu", kernel_initializer='glorot_normal'), Dropout(0.2))
model.add(BatchNormalization(momentum = 0.95, 
                             epsilon = 0.005, 
                             beta_initializer=RandomNormal(mean=0.0, stddev=0.05), 
                             gamma_initializer=Constant(value=0.9)))

model.add(Dense(100, activation="softmax"))

## Model summary
model.summary()
print(len(model.layers))

## Compile the model
model.compile(
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
    metrics=["accuracy"],
    optimizer=Adam(learning_rate=0.001, epsilon=1e-09)
)
```

```python
import tensorflow as tf
from tensorflow.keras.datasets import cifar100
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.layers import Dense, Flatten, GlobalAveragePooling2D
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam

# Cargar los datos CIFAR-100
(x_train, y_train), (x_test, y_test) = cifar100.load_data()

# Normalizar los datos
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# Convertir las etiquetas a one-hot encoding
y_train = to_categorical(y_train, 100)
y_test = to_categorical(y_test, 100)
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.2, random_state=42)

```

```python
# Cargar el modelo ResNet-50 preentrenado, excluyendo las capas superiores
base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(32, 32, 3))

# A√±adir capas superiores personalizadas para CIFAR-100
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(1024, activation='relu')(x)
predictions = Dense(100, activation='softmax')(x)

# Crear el modelo final
model_Resnet = Model(inputs=base_model.input, outputs=predictions)

# Congelar las capas base (opcional, si se desea un fine-tuning)
for layer in base_model.layers:
    layer.trainable = True

# Compilar el modelo
model_Resnet.compile(optimizer=Adam(learning_rate=0.0001), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Entrenar el modelo
history_Resnet = model_Resnet.fit(x_train, y_train, batch_size=128, epochs=300, validation_data=(x_val, y_val))
```
